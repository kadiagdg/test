```python
import os

# Chemins des fichiers CSV
fichier_mt100 = r"data\1224\extracted_data_mt100.csv"
fichier_mt900 = r"data\1224\extracted_data_mt900.csv"
fichier_rapport = r"data\1224\rapport_correspondance.txt"

# Fonction pour lire les références d'un fichier CSV
def lire_references(fichier_csv):
    references = set()
    if os.path.exists(fichier_csv):
        with open(fichier_csv, mode='r', encoding='utf-8') as csv_file:
            for ligne in csv_file.readlines()[1:]:  # Ignorer l'en-tête
                reference = ligne.split(';')[0]  # La référence est la première colonne
                references.add(reference)
    return references

# Lire les références des fichiers MT100 et MT900
references_mt100 = lire_references(fichier_mt100)
references_mt900 = lire_references(fichier_mt900)

# Trouver les correspondances et les différences
correspondances = references_mt100.intersection(references_mt900)
uniques_mt100 = references_mt100 - references_mt900
uniques_mt900 = references_mt900 - references_mt100

# Générer les statistiques
total_correspondances = len(correspondances)
total_uniques_mt100 = len(uniques_mt100)
total_uniques_mt900 = len(uniques_mt900)

# Exporter les résultats dans un fichier de rapport
with open(fichier_rapport, mode='w', encoding='utf-8') as rapport:
    rapport.write("Rapport de Correspondance entre MT1 et MT9\n")
    rapport.write("==============================================\n")
    rapport.write(f"Nombre total de références communes : {total_correspondances}\n")
    rapport.write(f"Nombre de références uniques à MT100 : {total_uniques_mt100}\n")
    rapport.write(f"Nombre de références uniques à MT900 : {total_uniques_mt900}\n")
    rapport.write("\n")

    # Détails des correspondances
    rapport.write("Références communes :\n")
    for ref in correspondances:
        rapport.write(f"{ref}\n")
    rapport.write("\n")

    # Détails des références uniques à MT100
    rapport.write("Références uniques à MT100 :\n")
    for ref in uniques_mt100:
        rapport.write(f"{ref}\n")
    rapport.write("\n")

    # Détails des références uniques à MT900
    rapport.write("Références uniques à MT900 :\n")
    for ref in uniques_mt900:
        rapport.write(f"{ref}\n")

print(f"Le rapport de correspondance a été généré dans {fichier_rapport}")
```
