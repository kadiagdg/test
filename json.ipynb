```python
import pandas as pd
import os

# Spécifiez le chemin vers votre fichier usage.log
fichier_log_path = 'exemple.log'

# Spécifiez le nombre de sous-fichiers souhaité
nombre_sous_fichiers = 10

# Créez un répertoire pour stocker les sous-fichiers si n'existe pas
output_directory = 'sous_fichiers'
os.makedirs(output_directory, exist_ok=True)

# Chargez le fichier log en DataFrame
df = pd.read_csv(fichier_log_path, delimiter='\t')

# Calculez la taille de chaque sous-fichier
taille_sous_fichier = len(df) // nombre_sous_fichiers

# Divisez le DataFrame en 10 sous-DataFrames
sous_dataframes = [df.iloc[i:i+taille_sous_fichier] for i in range(0, len(df), taille_sous_fichier)]

# Écrivez chaque sous-DataFrame dans un sous-fichier CSV
for i, sous_dataframe in enumerate(sous_dataframes):
    sous_fichier_path = os.path.join(output_directory, f'sous_fichier_{i + 1}.csv')
    sous_dataframe.to_csv(sous_fichier_path, index=False)

print(f'Le fichier a été divisé en {nombre_sous_fichiers} sous-fichiers.')
```

```python
import os

# Spécifiez le chemin vers votre fichier usage.log
fichier_log_path = 'exemple.log'

# Spécifiez le nombre de lignes par sous-fichier
lignes_par_sous_fichier = 10

# Créez un répertoire pour stocker les sous-fichiers si n'existe pas
output_directory = 'sous_fichiers'
os.makedirs(output_directory, exist_ok=True)

# Lecture du fichier log initial
with open(fichier_log_path, 'r') as file:
    lignes = file.readlines()

# Divisez le fichier en sous-fichiers de 10 lignes
for i in range(0, len(lignes), lignes_par_sous_fichier):
    sous_fichier_path = os.path.join(output_directory, f'sous_fichier_{i // lignes_par_sous_fichier + 1}.log')
    with open(sous_fichier_path, 'w') as sous_fichier:
        sous_fichier.writelines(lignes[i:i + lignes_par_sous_fichier])

print(f'Le fichier a été divisé en sous-fichiers de {lignes_par_sous_fichier} lignes.')
```

```python
import pandas as pd
import json
import os
from tqdm import tqdm
n=0
# Spécifiez le chemin vers le répertoire contenant les sous-fichiers
sous_fichiers_directory = 'sous_fichiers'

# Créez un DataFrame vide pour stocker les données
df_global = pd.DataFrame()

# Parcourez tous les sous-fichiers dans le répertoire
for sous_fichier_name in tqdm(os.listdir(sous_fichiers_directory)):
    sous_fichier_path = os.path.join(sous_fichiers_directory, sous_fichier_name)
    
    # Lire le contenu du sous-fichier
    with open(sous_fichier_path, 'r') as sous_fichier:
        lignes = sous_fichier.readlines()
    
    # Convertir chaque ligne JSON en dictionnaire et ajouter à une liste
    data_list = []
    for ligne in lignes:
        n=n+1
        log_data = json.loads(ligne)
        print(n)
        data_list.append(log_data)
    
    # Créer un DataFrame à partir de la liste de dictionnaires
    df_sous_fichier = pd.DataFrame(data_list)
    
    # Ajouter le DataFrame du sous-fichier au DataFrame global
    df_global = pd.concat([df_global, df_sous_fichier], ignore_index=True)

# Sauvegarder le DataFrame global en fichier parquet
df_global.to_parquet('resultat_global.parquet', index=False)

# Afficher les 5 premières lignes du DataFrame global
print(df_global.head())
```
